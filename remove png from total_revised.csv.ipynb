```python
file = pd.read_csv("total_revised.csv")
jewels = ["cushion", "emerald", "heart", "marquise", "oval", "pear", "princess", "round"]
```

```python
filenames = []
for jewel in jewels:
    filenames += os.listdir("C:/Users/user/Desktop/YAI/토이프로젝트/Diamonds2/images/" + jewel)
print(filenames)
```

```python
png = set()
for name in filenames:
    tmp = name.split('.')
    if tmp[1] == "png":
        png.add(tmp[0])
print(png)
```

```python
data = file.to_numpy()
```

```python
data = list(data)
```

```python
for i in range(len(data) - 1, -1, -1):
    if data[i][0] in png:
        data.pop(i)
```

```python
for i in data:
    output.write(",".join(map(str, i)) + "\n")
```

```python
output.close()
```
